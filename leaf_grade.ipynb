## Import all the libraried that required

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import time
```

## Load the pretrained yolov5 model

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
img = "kenguru-wheelchair-accessible-electric-vehicle.webp" # image 
```

```python
results = model(img) # check the prediction or parbability for the given image
results.print()
```

```python
# result iun images
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```
